# fattoriali di una lista

```python
lista = [0] * 3
fattoriali = [0] * 3

print("Dammi 3 valori di cui vuoi fare il fattoriale:")
for i in range(3):
    lista[i] = int(input("Inserisci un numero: "))

# Calcolo dei fattoriali
for i in range(3):
    risultato = 1
    for k in range(1, lista[i] + 1):
        risultato *= k
    fattoriali[i] = risultato  # Assegna il risultato finale solo qui
    print(f"Il fattoriale di {lista[i]} è: {fattoriali[i]}")
```

# quadrati di una lista

```python
lista = [0] * 3
quadrati = [0] * 3

for i in range(3):
    lista[i] = int(input("Inserisci un numero: "))

print("Calcolo il quadrato di tutti gli elementi della lista...")
for i in range(3):
    quadrati[i] = lista[i] ** 2
    print(f"Il quadrato di {lista[i]} è {quadrati[i]}")
```

# prodotto scalare

```python
lista1 = [0] * 3
lista2 = [0] * 3

risultato = 0
print("Inserisci le componenti dei due vettori:")

for i in range(3):
    lista1[i] = int(input(f"Inserisci il valore {i + 1} per il primo vettore: "))
    lista2[i] = int(input(f"Inserisci il valore {i + 1} per il secondo vettore: "))

for i in range(3):
    risultato += lista1[i] * lista2[i]

print("Il risultato del prodotto scalare è:", risultato)
```

# controllo di numeri

```python
from math import fabs

print("ciao dammi un numero")
numero = int(input())

if numero > 0:
    print("il numero è maggiore di 0")
elif numero == 0:
    print("il numero è 0")
else:
    print("il numero è minore di 0")

# Controllo di parità usando fabs
if fabs(numero) % 2 == 0:
    print("il numero è pari")
else:
    print("il numero è dispari")
```
